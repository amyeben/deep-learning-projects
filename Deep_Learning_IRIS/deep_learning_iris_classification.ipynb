Nous commençons par charger les données csv IRIS. Il s'agit de différentes types de fleurs de type Iris.

```python
data = pd.read_csv('Iris.csv', header=None)

# 1. Charger les données
df = pd.read_csv('iris.csv')
```

Afficher les catégories uniques de la colonne 'Species' : 

```python
print(df['Species'].unique())
```

```python
# 2. Explorer les données
print(df.head())
```

```python
print(df.describe())
```

```python
print(df.info())
```

```python
# Suppression de la première colonne par son nom
df = df.drop(columns=['Id'])
```

```python
# Moyenne des caractéristiques par espèce
df.groupby('Species').mean()
```

```python
print(df.isnull().sum())
```

```python
# 3. Visualiser les données
# Distribution
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='SepalLengthCm', hue='Species', kde=True)
plt.title('Distribution de la longueur des sépales par espèce')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='SepalWidthCm', kde=True, hue='Species')
plt.title('Distribution de la largeur des sépales par espèce')
plt.show()
```

```python
# Corrélation
plt.figure(figsize=(10,8))
# Exclude the 'Species' column from correlation calculation
sns.heatmap(df.drop(columns='Species').corr(), annot=True, cmap='coolwarm')
plt.title('Matrice de corrélation des caractéristiques')
plt.show()
```

```python
sns.pairplot(df, hue='Species')
plt.suptitle('Pairplot des caractéristiques par espèce', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='SepalWidthCm', y='PetalWidthCm', hue='Species')
plt.title('Largeur des sépales vs. Largeur des pétales')
plt.show()
```
